```python
df = pd.read_csv('data/cleaned_data.csv')
```

```python
df.head()
```

```python
df.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(labels=['Day','Month','Year','Order_Hours',
                'Order_Min','Picked_Hours','Picked_Min'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Independent and Dependaent Variablew

X = df.drop(['Time_taken_min'],axis=1)
y=df['Time_taken_min']
```

```python
X.head()
```

```python
y.head()
```

```python
#segregate Numerical and categorical features

categorical_cols = X.select_dtypes(include='O').columns
numerical_cols = X.select_dtypes(exclude='O').columns

print('Numerical columns:',numerical_cols)
print('categorical columns:',categorical_cols)
```

```python
#Defining custome ranking
whether_condition_map = ['Sunny','Stromy','Sandstroms','Windy','Fog','Cloudy']
Road_Traffic_map = ['Low','Medium','High','Jam']
Type_of_vehicle_map = ['bicycle','electric_scooter','motorcycle']
Festival_map=['No','Yes']
city_map = ['Urban','Metropolitian','Semi-Urban']
```

```python
from sklearn.impute import SimpleImputer #handle missing
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
#pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

#OrdinalEncoder(categories=[whether_condition_map,Road_Traffic_map,Type_of_vehicle_map,Festival_map,city_map])
```

```python
#num pipeline

num_pipeline = Pipeline(
   steps=[
      ('impute',SimpleImputer(strategy='median')),
      ('scaler',StandardScaler())
   ]
)
#cat pipeline
cat_pipeline = Pipeline(
   steps=[
      ('impute',SimpleImputer(strategy='most_frequent')),
      ('oneHot',OneHotEncoder(sparse_output=False)),
      ('scaler',StandardScaler())
   ]
)

preprocessor = ColumnTransformer(
   [('num_pipeline',num_pipeline,numerical_cols),
    ('cat_pipeline',cat_pipeline,categorical_cols)]
)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.head()
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression,ElasticNet,Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
#def evalute model

def evalute_model(true,pred):
   mse = mean_squared_error(true,pred)
   rmse=np.sqrt(mse)
   mae = mean_absolute_error(true,pred)
   r2_square= r2_score(true,pred)
   return rmse,mae,r2_square
```

```python
#Train Multple model

models = {
   'LinearRegression':LinearRegression(),
   'Ridge':Ridge(),
   'Lasso':Lasso(),
   'ElasticNet':ElasticNet(),
   "DecisionTree":DecisionTreeRegressor(),
   'RandomForest':RandomForestRegressor(),
   'Neighbours':KNeighborsRegressor(),
   'Svm':SVR()
}

model_list = []
r2_list=[]

for i in range(len(models)):
   model =list(models.values())[i]
   model.fit(X_train,y_train)

   #make prediction
   y_pred = model.predict(X_test)
   rmse,mae, r2_square=evalute_model(y_test,y_pred)

   print(list(models.keys())[i])
   model_list.append(list(models.keys())[i])
   print('Model Training Performance')
   print('RMSE:',rmse)
   print('MAE:',mae)
   print('R2_SCore:',r2_square*100)

   r2_list.append(r2_square)

   print('='*35)
   print('\n')
```
